```python
import networkx as nx 
import numpy as np
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()

G.add_node(1)
G.add_node(2)
G.add_node(3)
G.add_node(4)
G.add_node(5)
G.add_node(6)

G.add_edges_from([(3, 2), (3, 5), (1, 2), (1, 4), (5, 4),(6, 4), (6, 3), (1, 6) ])

# Define positions for the nodes (in this case, a simple linear layout)
pos = {1: (0, 1), 2: (-0.5, 0), 3: (-0.3, -1), 4: (0.3, 0.3), 5: (0.4, -0.6),6: (0, 0)}

# Draw the edges
nx.draw_networkx_edges(G, pos)
# Draw the nodes
nx.draw_networkx_nodes(G, pos)
# Add labels if desired
nx.draw_networkx_labels(G, pos)
# Display the graph
plt.show()
```

```python
am = nx.adjacency_matrix(G)
am = am.todense()
am
```
